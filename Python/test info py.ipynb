```python
def explode(a):
    ex=[]
    h=0.1
    for i in range (1,a+1):
        ex.append(h)
    return tuple(ex)

def getList(dict):
    return dict.keys()
```

```python
pie1=data.fuel_type.value_counts()
dict=pie1.to_dict()
col=sns.color_palette('Set2_r')
plt.pie(pie1,labels=getList(dict),explode=explode(len(pie1)),radius=1,autopct="%1.1f%%",shadow=True,startangle=90,colors=col);
```

```python
pie2=data.sales_channel.value_counts()
dict=pie2.to_dict()
col1=sns.color_palette('icefire')
plt.pie(pie2,labels=getList(dict),explode=explode(len(pie2)),radius=1,autopct="%1.1f%%",shadow=True,startangle=90,colors=col1);
```
